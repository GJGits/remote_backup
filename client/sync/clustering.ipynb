```python
# ---------- IMPORT ---------- #
import os
import pandas
from scipy.interpolate import CubicSpline
from sklearn.cluster import SpectralClustering
import numpy as np
from scipy.signal import periodogram
import matplotlib.pyplot as plt
```

```python
# ----- CONSTANTS ----- #
in_dir = "datasets/features_post_avgM"
out_dir = ""
features = ["maxHR", "MaxRR", "StdHR", "lowFrequencyPower/highFrequencyPower", "%lowFrequencyPower", "VeryHighFrequenciesPower", "PNN50", "HighFrequenciesPower", "StdRR"]
subjectsToRemove = ["01", "02", "40"]
```

```python
# ----- FUNCTIONS ----- #
```

```python
# ----- MAIN ----- #
samples = []
for filename in os.listdir(in_dir):
    paziente = "{}".format(filename[1:3])
    if filename == "README.txt" or paziente in subjectsToRemove:
        continue
    featuresDF = pandas.read_csv("{}/{}".format(in_dir, filename), sep='\t', usecols = features ,low_memory=False)
    samples.append(featuresDF.mean(axis=0).to_list())
samples = np.array(samples)
for i in range(10):
    print("random state:{}".format(i))
    print("=====")
    for j in [0.1, 0.2, 0.5]:
        clustering = SpectralClustering(n_clusters=3, gamma=j, assign_labels="discretize", random_state=i).fit(samples)
        print("gamma={}:".format(j))
        print(clustering.labels_)
    print("=====")
```
